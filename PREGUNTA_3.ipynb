Instalando LLAMA

```python
#@title Cargando LLAMA3
import transformers
import torch

model_id = "unsloth/llama-3-8b-Instruct-bnb-4bit"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={
        "torch_dtype": torch.float16,
        "quantization_config": {"load_in_4bit": True},
        "low_cpu_mem_usage": True,
    },
)
```

##Cargando el CSV con los abstracts a trabajar

```python
import pandas as pd

# Enlace al archivo CSV en GitHub
url = 'https://raw.githubusercontent.com/AHINTOR/PARCIAL_cuadernos_python/main/scopus_vpp.csv'

# Leer el archivo CSV desde el enlace, guardo como ds
ds = pd.read_csv(url)
```

```python
# de mi tabla en ds voy a guardar solamente los articulos 5 en una nueva
# variable df
df = ds.iloc[0:5]
```

```python
df
```

```python
def consultas(input):
  messages = [
    {"role"   : "system",
     "content": context1},
    {"role"   : "user",
     "content": questions},
  ]

  prompt = pipeline.tokenizer.apply_chat_template(
          messages,
          tokenize=False,
          add_generation_prompt=True
  )

  terminators = [
      pipeline.tokenizer.eos_token_id,
      pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = pipeline(
      prompt,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=True,
      temperature=0.6,
      top_p=0.9,
  )

  return outputs[0]["generated_text"][len(prompt):]
```

##Código de bucle para concatenar la lista de preguntas creada previamente con los abstracts

```python
df['Analisis1'] = None

for i in range(5):
  context1 = df["Abstract"][i]
  response = consultas(context1)
  df['Analisis1'][i] = response
```
