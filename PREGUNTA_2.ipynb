# (b) Usando matplotlib y/o seaborn grafique un diagrama de barras con los tipos de publicación versus la cantidad.

```python
import pandas as pd

# Enlace al archivo CSV en GitHub
url = 'https://raw.githubusercontent.com/AHINTOR/PARCIAL_cuadernos_python/main/scopus_vpp.csv'

# Leer el archivo CSV desde el enlace
df = pd.read_csv(url)

# Mostrar las primeras filas del archivo
df.head()

```

Análisis descriptivo

```python
# Mostrar información general del DataFrame (tipos de datos, valores nulos, etc.)
df.info()
```

```python
# Mostrar los nombres de las columnas
print('Nombre columnas:', df.columns)
```

```python
# Mostrar la cantidad de filas y columnas
print("Cantidad de Filas y Columnas:", df.shape)
```

## Creando el grafico con matplotlib



```python
import matplotlib.pyplot as plt

# Contar la cantidad de publicaciones por tipo de documento (Document Type)
doc_type_counts = df['Document Type'].value_counts()

# Crear un diagrama de barras con los tipos de publicación y la cantidad
plt.figure(figsize=(10, 6))
doc_type_counts.plot(kind='bar', color='skyblue')

# Agregar etiquetas y título
plt.title('Tipos de Publicación vs Cantidad', fontsize=14)
plt.xlabel('Tipos de Publicación', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)

# Mostrar el gráfico
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
